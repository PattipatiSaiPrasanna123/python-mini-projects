```python
import pandas as pd
```

```python
Datasetpath='/content/drive/MyDrive/retail_transactions.csv'
```

```python
df=pd.read_csv(Datasetpath)
```

```python
df
```

```python

# Calculate total revenue
df["revenue"] = df["Quantity"] * df["Price"]
total_revenue = df["revenue"].sum()
```

```python
print(f"Total Revenue: ${total_revenue:,.2f}")
```

```python

most_frequent_category = df['ProductCategory'].value_counts().mode()
```

```python
print("Most frequently purchased Category:", most_frequent_category)
```

```python

average_transcation_amount = df['revenue'].mean()
print(f"Average Transaction Amount: ₹{average_transcation_amount:,.2f}")
```
