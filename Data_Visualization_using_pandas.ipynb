```python
scorebybrazil = df['home_team'] =='Brazil'
```

```python
brazildf = df[ScoreAboveZero & scorebybrazil].sort_values(by = 'home_score', ascending = True)
```

```python
brazildf.info()
```

```python
%matplotlib inline
```

```python
from matplotlib import pyplot as plt
```

```python
brazildf.plot(x='home_score', y='away_score', kind='scatter')
plt.xlabel('home score')
plt.ylabel('score')
plt.legend('Takeaway')
plt.title('Brazil Matches')
plt.show()
```

```python

brazildf.plot(x='Year',y='home_score',kind='hexbin')
plt.xlabel('Year')
plt.ylabel('Home Score')
plt.legend('Takeaway')
plt.title('Brazil Matches')
plt.show()
```
