```python
def easy_quiz():
    score = 0
    
    # Question 1
    print("Question 1: What is 2 / 2?")
    answer1 = input("a) 3 b) 1 c) 5 d) 6 Your answer: ")

    if answer1 == "b":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 1.")
    
    # Question 2
    print("Question 2: What is 6 / 2?")
    answer2 = input("a) 10 b) 9  c) 3 d) 13 Your answer: ")
    
    if answer2 == "c":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 3.")

        # Question 3
    print("Question 3: What is 10 / 5?")
    answer3 = input("a) 6 b) 7 c) 2 d) 9 Your answer: ")
    
    if answer3 == "c":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 2.")

        # Question 4
    print("Question 4: What is 20/ 5?")
    answer4 = input("a) 4 b) 17 c) 15 d) 9 Your answer: ")
    
    if answer4 == "a":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 4.")

        # Question 5
    print("Question 5: What is 30 / 15?")
    answer5 = input("a) 4 b) 7 c) 2 d) 19 Your answer: ")
    
    if answer5 == "c":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 2.")

        # Question 6
    print("Question 6: What is 24 / 2?")
    answer6 = input("a) 15 b) 12 c) 10 d) 36 Your answer: ")
    
    if answer6 == "b":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 12.")

        # Question 7
    print("Question 7: What is 25 / 5?")
    answer7 = input("a) 3 b) 2 c) 8 d) 5 Your answer: ")
    
    if answer7 == "d":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 5.")
    
    # Question 8
    print("Question 8: What is 50 / 30?")
    answer8 = input("a) 6.0 b) 7 c) 1.66 d) 3 Your answer: ")
    
    if answer8 == "c":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 1.66.")

        # Question 9
    print("Question 9: What is 156 / 40?")
    answer9 = input("a) 3.9 b) 1.9 c) 1.8 d) 3.8 Your answer: ")
    
    if answer9 == "a":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 3.9.")

        # Question 10
    print("Question 10: What is 50 / 5?")
    answer10 = input("a) 6 b) 10 c) 4 d) 12 Your answer: ")
    
    if answer10 == "b":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 10.")
    
    
    # Final score
    print(f"Your final score is: {score}/100")

# Run the quiz
easy_quiz()
```
